````python
# Calculate the coefficient of best fit line or curve
x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100,10,32,49,10,10]
y = [88,12,23,3,6,8,9,12,27,36,34,77,60,35,36,37,18,49,50,100,10,32,49,10,10,22]

print(np.polyfit(x,y,1))
np.corrcoef(x,y)
````

````python
# sample points 
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
    '''
    Return m, b
    '''
    #set up lists for x^2 and xy
    X_sqr = []
    XY = []

    #fill lists
    for i in range(len(X)):
        X_sqr.append(X[i] ** 2)
        XY.append(X[i]*Y[i])
    
    x_s = sum(X)
    y_s = sum(Y)
    x_sqr_s = sum(X_sqr)
    xy_s = sum(XY)
    
    n = len(X)
    
    m = (n*xy_s-x_s*y_s)/(n*x_sqr_s-x_s **2)
    b = (y_s - m*x_s)/n
    
    y_lin = []
    
    for i in X:
        y_lin.append(m*i+b)
    
    
    print('best fit line:\ny = {:.2f} + {:.2f}x'.format(b,m))
    
    return m, b, y_lin

# solution
m, b, y_lin = best_fit(X, Y)
#best fit line:
#y = 0.80 + 0.92x

# plot points and fit line
plt.style.use('seaborn')
plt.plot(X,Y,'o')
plt.plot(X,y_lin)
plt.show()
print(np.polyfit(X,Y,1))
````
